```python
import pandas as pd
```

```python
file_path = '/content/drive/MyDrive/Concept Ai/week 3/Titanic-Dataset (1).csv'
titanic = pd.read_csv(file_path)
```

```python
titanic_info = titanic.info()
```

```python
#dataframe containing only the 'Fare' col
fare = titanic[['Fare']]
print("Fare head:\n",fare.head())
```

```python
#dataframe containg 'Pclass' and 'Age' columns
class_age = titanic[['Pclass','Age']]
print("Class and Age head:\n",class_age.head())
```

```python
#dataFrame containing 'Survived' and 'Sex' col
survived_gender = titanic[['Survived','Sex']]
print("Survived and Sex head:\n",survived_gender.head())
```

```python
#passengers with fare greater than 100
fare_gt_100 = titanic[titanic['Fare']>100]
print("Passengers with fare > 100:\n",fare_gt_100)
```

```python
#passengers in first class(Pclass = 1)
first_class = titanic[titanic['Pclass'] == 1]
print("First class passengers: \n",first_class)
```

```python
#passengers under 18
female_under_18 = titanic[(titanic['Age']<18 &(titanic['Sex']=='female'))]
print("Femake passengers under 18:\n",female_under_18)
```

```python
#subsetting rows by categorical variables

#passengers who embarked from 'C' or 'S'
embarked_c_or_s  = titanic[titanic['Embarked'].isin(['C','S'])]
print("Passengers in first or second class:\n",embarked_c_or_s.head())
```

```python
#passengers in first or second class(Pclass in[1,2])
first_second_class = titanic[titanic['Pclass'].isin([1,2])]
print("Passengers in first or second class:\n",first_second_class.head())
```

```python
#fare relative age analysis
#handle missing age values by filling them with the median
median_age = titanic['Age'].median()
titanic['Age'].fillna(median_age,inplace =True)
```

```python
#Add col 'fare divided by age
titanic['fare_per_year'] = titanic ['Fare']/titanic['Age']
```

```python

# Subset rows where fare_per_year is greater than 5
high_fare_age = titanic[titanic['fare_per_year'] > 5]
```

```python
#  Sort high_fare_age by descending fare_per_year
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)
```

```python
# Select 'Name' and 'fare_per_year' columns and save as 'result'
result = high_fare_age_srt[['Name', 'fare_per_year']]
print("Highest fare per year passengers:\n", result.head())
```

```python

#  Add a column 'fare_per_class' (Fare divided by Pclass)
titanic['fare_per_class'] = titanic['Fare'] / titanic['Pclass']

#  Subset for adult males (Age >= 18 and Sex = 'male')
adult_males = titanic[(titanic['Age'] >= 18) & (titanic['Sex'] == 'male')]

#  Sort adult males by descending 'fare_per_class'
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)

#  Select 'Name', 'Age', and 'fare_per_class' columns and save as 'result'
result = adult_males_srt[['Name', 'Age', 'fare_per_class']]
print("Highest fare per class for adult males:\n", result.head())
```

```python

# 1. Calculate the total Fare paid across all passengers
total_fare = titanic['Fare'].sum()

# 2. Subset for passengers in first class (Pclass is 1) and calculate their total fare
fare_first_class = titanic[titanic['Pclass'] == 1]['Fare'].sum()

# 3. Subset for passengers in second class (Pclass is 2) and calculate their total fare
fare_second_class = titanic[titanic['Pclass'] == 2]['Fare'].sum()

# 4. Subset for passengers in third class (Pclass is 3) and calculate their total fare
fare_third_class = titanic[titanic['Pclass'] == 3]['Fare'].sum()

# 5. Combine the fare totals from first, second, and third classes into a list
fare_totals = [fare_first_class, fare_second_class, fare_third_class]

# 6. Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class
fare_proportions = [(fare / total_fare) * 100 for fare in fare_totals]
print("Fare proportions for each class as a percentage:\n", fare_proportions)

```

```python

# 1. Create a new column, 'age_group', categorizing passengers into 'child', 'adult', 'senior'
# Categories: Child (<18), Adult (18-64), Senior (65+)
titanic['age_group'] = pd.cut(titanic['Age'], bins=[0, 17, 64, 120], labels=['child', 'adult', 'senior'])

# 2. Calculate the total number of passengers on the Titanic
total_passengers = len(titanic)

# 3. Count the number of passengers in each age group
age_group_counts = titanic['age_group'].value_counts()

# 4. Divide the count of each age group by the total number of passengers to get the proportion of passengers in each age group
age_group_proportions = (age_group_counts / total_passengers) * 100

# 5. Display the proportion as a percentage
print("\nProportion of passengers in each age group as a percentage:\n", age_group_proportions)
```
